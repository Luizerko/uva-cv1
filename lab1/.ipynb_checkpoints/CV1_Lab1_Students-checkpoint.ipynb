### 1.1 RGB Colour Model (2 pts)

Why do we use a RGB colour model as the basis of our digital cameras and photography? How does a standard digital camera capture the full RGB colour image?


We use the RGB model as the basis for our cameras and digital photographs because of its analogy with the human imaging sensory system. We humans perceive colors through the use of cells called cones at the back of our eyes. These cells are especially sensitive to the colors red, green and blue, and build the perception of other colors primarily from combinations of these three spectra. The use of the RGB model for digital image representation, therefore, is nothing more than the use of a system that is extremely similar to the natural functioning of human color perception.

A standard digital camera works in a few steps. The equipment first lets in light through a small hole in the front. This light then passes through a lens to be properly directed. The lens has various functions related to light intensity and image focus. Once properly directed inside the equipment, the light hits photosensitive tiles that will record the intensity of the light at that point in the moment the photo is taken. Note that it is at this last stage that the RGB model is used, since there are three photosensitive tiles used, one for red intensities, another for green and a final one for blue.

````python
def normalize_axis(array):
    return (array/np.max(np.abs(array)) + 1)/2
````

````python
def rgb2grays(input_image, method='opencv'):
    
    if method == 'lightness':
        def min_max_average(rgb):
            return int(max(rgb)/2+min(rgb)/2)
        new_image = np.apply_along_axis(min_max_average, 2, image)
    
    elif method == 'average':
        def average(rgb):
            return int(rgb[0]/3 + rgb[1]/3 + rgb[2]/3)
        new_image = np.apply_along_axis(average, 2, image)
        
    elif method == 'luminosity':
        def luminosity(rgb):
            return int(0.21*rgb[0] + 0.72*rgb[1] + 0.07*rgb[2])
        new_image = np.apply_along_axis(luminosity, 2, image)
    
    elif method == 'opencv':
        new_image = cv2.cvtColor(input_image, cv2.COLOR_RGB2GRAY)
    
    return new_image


def rgb2opponent(input_image):
    def opponent(rgb):
        #rgb_aux = np.array(rgb, dtype=np.int)
        c = []
        c.append((rgb[0] - rgb[1])/np.sqrt(2))
        c.append((rgb[0] + rgb[1] - 2*rgb[2])/np.sqrt(6))
        c.append((rgb[0] + rgb[1] + rgb[2])/np.sqrt(3))
        return np.array(c)
    new_image = np.apply_along_axis(opponent, 2, image)
    
    # Normalizing to make it better to see on imshow and avoid clipping
    for i in range(3):
        new_image[:, :, i] = normalize_axis(new_image[:, :, i])
    return new_image

def rgb2normedrgb(input_image):
    def normedrgb(rgb):
        #rgb_aux = np.array(rgb, dtype=np.int)
        c = [i/sum(rgb) for i in rgb]
        return np.array(c)
    new_image = np.apply_along_axis(normedrgb, 2, image)
    return new_image
````

````python
def ConvertColourSpace(input_image, colourspace):
    '''
    Converts an RGB image into a specified color space, visualizes the
    color channels and returns the image in its new color space.

    Colorspace options:
      opponent
      rgb -> for normalized RGB
      hsv
      ycbcr
      gray

    P.S: Do not forget the visualization part!
    '''

    # Convert the image into double precision for conversions
    input_image = input_image.astype(np.float32)

    if colourspace.lower() == 'opponent':
        new_image = rgb2opponent(input_image)

    elif colourspace.lower() == 'rgb':
        new_image = rgb2normedrgb(input_image)

    elif colourspace.lower() == 'hsv':
        new_image = cv2.cvtColor(input_image, cv2.COLOR_RGB2HSV)
        
        # Normalizing to make it better to see on imshow and avoid clipping
        for i in range(3):
            new_image[:, :, i] = normalize_axis(new_image[:, :, i])

    elif colourspace.lower() == 'ycbcr':
        new_image = cv2.cvtColor(input_image, cv2.COLOR_RGB2YCR_CB)
        
        # Normalizing to make it better to see on imshow and avoid clipping
        for i in range(3):
            new_image[:, :, i] = normalize_axis(new_image[:, :, i])
            
    elif colourspace.lower() == 'gray':
        new_image = []
        new_image.append(rgb2grays(input_image, 'lightness'))
        new_image.append(rgb2grays(input_image, 'average'))
        new_image.append(rgb2grays(input_image, 'luminosity'))
        new_image.append(rgb2grays(input_image))
    
    else:
        print('Error: Unknown colorspace type [%s]...' % colourspace)
        new_image = input_image

    return new_image
````

### 1.3 Colour Space Properties (2 pts)

Explain each of the above 5 colour spaces and their properties. What are the benefits of using a different colour space other than RGB? Provide reasons for each of the above cases. You can include your observations from the visualizations.

Gray colorspace: \
This is a one channel colorspace in which pixels are represented by their luminosity, from black to white. Black pixels are the ones with less luminosity and white pixels are the ones with more luminosity. \
The major benefit of using grayscale is its data size reduction. Not only it makes the image much less expensive to store, but also makes it easier to process - on a neural network for example.

Opponent colorspace: \
This is a colorspace based on the human vision. It contains one channel for perceptual lightness and two channels that vary in colors that the human eye sees uniquely - green to red and blue to yellow. \
The idea of this space, and its biggest benefit, is to provide a better mapping between the values changes on the coordinates and the change of colors for human eyes, thus making it easier for humans to work with.

Normalized RGB colorspace: \
This colorspace is very similar to the normal RGB, but with much lower values since each channel is normalized by the sum of all the channels. \
The biggest benefit of this colorspace comes directly because of the normalization on the ntensity of all the channels for each pixel and it is providing an image representation with less light influence, so more focused on the color information. The normed RGB image provided above makes it clear.

HSV colorspace: \
As seen in class, this colorspace represents the color of the pixel (H), the saturation of the color from white to its strongest version (S) and the color brightness from black to its brightest version (V). \
The bigges benefit of this colorspace is to make color and light information separable, making it easier for humans to work with colors without being misled by color perception because of the light. The Hue channel image provided above makes the color separation clear.

YCbCr colorspace: \
This colorspace is similar to the opponent colorspace in the sense that it has one channel more focused on perceptual lightness and two channels focused on color. This colorspace, however, is not focused on human perception, but on optimization. Therefore, instead of using the green-red and blue-yellow channels, it makes use of the difference between the blue color and a reference value and the difference between the red color and another reference value. \
The benefits of this colorspace are mostly related to data compression. Abusing the fact that human eyes are more sensitive to luminance than to colors, one can reduce the amount of stored information on the two color channels without generating too much perceived loss on the image quality.

````python
# Helper functions to load the synthetic images

def load_syn_images(image_dir='./images/photometrics_images/SphereGray5/', channel=0):
    files = os.listdir(image_dir)
    nfiles = len(files)

    image_stack = None
    V = 0
    Z = 0.5

    for i in range(nfiles):
        # read input image
        im = cv2.imread(os.path.join(image_dir, files[i]))
        im = np.flip(im, axis=-1)
        im = im[:,:,channel]

        # stack at third dimension
        if image_stack is None:
            h, w = im.shape
            print('Image size (H*W): %d*%d' %(h,w) )
            image_stack = np.zeros([h, w, nfiles], dtype=int)
            V = np.zeros([nfiles, 3], dtype=np.float64)

        image_stack[:,:,i] = im

        # read light direction from image name
        X = np.double(files[i][(files[i].find('_')+1):files[i].rfind('_')])
        Y = np.double(files[i][files[i].rfind('_')+1:files[i].rfind('.png')])
        V[i, :] = [-X, Y, Z]

    # normalization
    image_stack = np.double(image_stack)
    min_val = np.min(image_stack)
    max_val = np.max(image_stack)
    image_stack = (image_stack - min_val) / (max_val - min_val) if max_val!=min_val else np.zeros(image_stack.shape)  # avoid fail when image_stack==0
    normV = np.tile(np.sqrt(np.sum(V ** 2, axis=1, keepdims=True)), (1, V.shape[1]))
    scriptV = V / normV

    return image_stack, scriptV
````
